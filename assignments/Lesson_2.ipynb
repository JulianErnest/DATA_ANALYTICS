## Dataset 

This dataset contains the hourly and daily count of rental bikes between years 2011 and 2012 in Capital bikeshare system with the corresponding weather and seasonal information.

```python
import pandas as pd

data = pd.read_csv("../Datasets/bike-sharing/day.csv")
data
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


# Convert 'dteday' column to datetime format
data['dteday'] = pd.to_datetime(data['dteday'])

# Plot 1: Line chart showing trend of total bike rentals over time
plt.figure(figsize=(10, 6))
sns.lineplot(x='dteday', y='cnt', hue='yr', data=data)
plt.title('Total Bike Rentals Over Time')
plt.xlabel('Date')
plt.ylabel('Total Count of Rentals')
plt.xticks(rotation=45)
plt.legend(title='Year')
plt.tight_layout()
plt.show()

# Plot 2: Bar chart showing distribution of bike rentals by season
plt.figure(figsize=(8, 6))
sns.barplot(x='season', y='cnt', data=data, estimator=sum, ci=None)
plt.title('Distribution of Bike Rentals by Season')
plt.xlabel('Season')
plt.ylabel('Total Count of Rentals')
plt.xticks(ticks=[0, 1, 2, 3], labels=['Spring', 'Summer', 'Fall', 'Winter'])
plt.tight_layout()
plt.show()

# Plot 3: Stacked area chart showing breakdown of bike rentals by weather situation
plt.figure(figsize=(10, 6))
sns.lineplot(x='dteday', y='cnt', hue='weathersit', data=data, estimator=sum, ci=None)
plt.title('Breakdown of Bike Rentals by Weather Situation')
plt.xlabel('Date')
plt.ylabel('Total Count of Rentals')
plt.xticks(rotation=45)
plt.legend(title='Weather Situation')
plt.tight_layout()
plt.show()

# Plot 4: Scatter plot showing relationship between temperature and total count of bike rentals
plt.figure(figsize=(8, 6))
sns.scatterplot(x='temp', y='cnt', hue='weathersit', data=data, palette='viridis')
plt.title('Relationship between Temperature and Bike Rentals')
plt.xlabel('Temperature (Celsius)')
plt.ylabel('Total Count of Rentals')
plt.tight_layout()
plt.show()

# Plot 5: Bubble chart showing relationship between humidity, wind speed, and total count of bike rentals
plt.figure(figsize=(10, 8))
sns.scatterplot(x='hum', y='windspeed', size='cnt', hue='season', sizes=(20, 200), data=data, palette='Set2', alpha=0.7)
plt.title('Relationship between Humidity, Wind Speed, and Bike Rentals')
plt.xlabel('Humidity')
plt.ylabel('Wind Speed')
plt.tight_layout()
plt.show()
```
