**TASK: Use DBSCAN and a for loop to create a variety of models testing different epsilon values. Set min_samples equal to 2 times the number of features. During the loop, keep track of and log the percentage of points that are outliers. For reference the solutions notebooks uses the following range of epsilon values for testing:**

    

```python
from sklearn.cluster import DBSCAN
import numpy as np

epsilon_values = np.linspace(0.001,3,50)

min_samples = 2 * scaled_wholesale_customers_df.shape[1]

outlier_percentages = []

for epsilon in epsilon_values:
    dbscan = DBSCAN(eps=epsilon, min_samples=min_samples)
    dbscan.fit(scaled_wholesale_customers_df)
    
    n_outliers = np.sum(dbscan.labels_ == -1)
    percentage_outliers = (n_outliers / len(scaled_wholesale_customers_df)) * 100
    
    outlier_percentages.append(percentage_outliers)
    
    print(f"Epsilon: {epsilon}, Percentage of outliers: {percentage_outliers:.2f}%")

# Plot the percentage of outliers for each epsilon value
plt.figure(figsize=(10, 6))
plt.plot(epsilon_values, outlier_percentages, marker='o', linestyle='-')
plt.title('Percentage of Outliers vs. Epsilon Values')
plt.xlabel('Epsilon Values')
plt.ylabel('Percentage of Outliers')
plt.grid(True)
plt.show()
```

**TASK: Create a line plot of the percentage of outlier points versus the epsilon value choice.**

```python
plt.figure(figsize=(10, 6))
plt.plot(epsilon_values, outlier_percentages, marker='o', linestyle='-')
plt.title('Percentage of Outlier Points vs. Epsilon Value Choice')
plt.xlabel('Epsilon Values')
plt.ylabel('Percentage of Outlier Points')
plt.grid(True)
plt.show()
```

**TASK: Compare the statistical mean of the clusters and outliers for the spending amounts on the categories.**

```python
clusters = df[df['Labels'] != -1].drop(columns=['Channel', 'Region'])
outliers = df[df['Labels'] == -1].drop(columns=['Channel', 'Region'])

cluster_mean = clusters.groupby('Labels').mean()
outlier_mean = outliers.groupby('Labels').mean()

compare_mean_df = pd.concat([cluster_mean, outlier_mean], keys=['Cluster', 'Outlier'])
compare_mean_df
```

**TASK: Normalize the dataframe from the previous task using MinMaxScaler so the spending means go from 0-1 and create a heatmap of the values.**

```python
from sklearn.preprocessing import MinMaxScaler

# Initialize the MinMaxScaler
scaler = MinMaxScaler()

# Fit and transform the dataframe to scale the values between 0 and 1
normalized_compare_mean_df = pd.DataFrame(scaler.fit_transform(compare_mean_df), 
                                          columns=compare_mean_df.columns, 
                                          index=compare_mean_df.index)

# Create a heatmap of the normalized values
plt.figure(figsize=(10, 8))
sns.heatmap(normalized_compare_mean_df, cmap='coolwarm', annot=True, fmt=".2f")
plt.title('Heatmap of Normalized Spending Means for Clusters and Outliers')
plt.xlabel('Features')
plt.ylabel('Cluster/Outlier')
plt.show()
```

**TASK: Create another heatmap similar to the one above, but with the outliers removed**

```python
normalized_cluster_mean_df = normalized_compare_mean_df.xs('Cluster', level=0)

# Create a heatmap of the normalized cluster means
plt.figure(figsize=(10, 8))
sns.heatmap(normalized_cluster_mean_df, cmap='coolwarm', annot=True, fmt=".2f")
plt.title('Heatmap of Normalized Spending Means for Clusters')
plt.xlabel('Features')
plt.ylabel('Cluster')
plt.show()
```

**TASK: What spending category were the two clusters mode different in?**

```python
# Calculate the difference in means between the two clusters
cluster_diff = normalized_cluster_mean_df.diff()

# Find the spending category with the largest difference in means
most_different_category = cluster_diff.abs().mean().idxmax()

print(f"The most different spending category between the two clusters is: {most_different_category}")
```
